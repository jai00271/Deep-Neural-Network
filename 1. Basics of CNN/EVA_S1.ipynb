<h1>Import packages<h1>

```
# import the necessary packages
import numpy as np
from urllib.request import urlopen
import cv2
from matplotlib import pyplot as plt
```

<h1> Read the file </h1>
 
 Here we are trying to achieve 3 things:
  

1.   download the image
2.   convert it to a NumPy array
3.   read it into OpenCV format



```
# METHOD #1: OpenCV, NumPy, and urllib
def url_to_image(url): 
	resp = urlopen(url)
	image = np.asarray(bytearray(resp.read()), dtype="uint8")
	image = cv2.imdecode(image, cv2.IMREAD_COLOR)
 
	# return the image
	return image
```

<h1> Load an image <h1>

```
image = url_to_image("https://fontsarena-cd5e.kxcdn.com/wp-content/uploads/2019/04/helvetica-now-font-400x364.png")
```

```
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```
edges = cv2.Canny(image,100,200)
```

```
cv2_imshow(edges)
```

Here we are using convolution of 3x3 which detects verticle edges.

```
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([[-1,0,1],[-1,0,1],[-1,0,1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Horizontal Edge Detector

Convolution to detect horizontal line: [[-1,-1,-1],[2,2,2],[-1,-1,-1]]

```
kernel = np.float32([[-1,-1,-1],[2,2,2],[-1,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

45 Degree Angle Detector

Convolution to detect 45 Degree Angle: [[-1,-1,2],[-1,2,-1],[2,-1,-1]]

```
kernel = np.float32([[-1,-1,2],[-1,2,-1],[2,-1,-1]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Blur Kernel

Convolution to detect Blur edges:[[0.0625,0.125,0.0625],[0.125,0.25,0.125],
[0.0625,0.125,0.0625]]

```
kernel = np.float32([[0.0625,0.125,0.0625],[0.125,0.25,0.125],[0.0625,0.125,0.0625]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Sharpen Kernel

Convolution to detect Sharp edges:
[[0,-1,0],[-1,5,-1],[0,-1,0]]

```
kernel = np.float32([[0,-1,0],[-1,5,-1],[0,-1,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```

Identity function (doesn't do anything)

Convolution to detect Sharp edges:
[[0,0,0],[0,1,0],[0,0,0]]

```
kernel = np.float32([[0,0,0],[0,1,0],[0,0,0]])

dst = cv2.filter2D(image,-1,kernel)
cv2_imshow(dst)
```
